## 1. Set / Himpunan

- Himpunan (set) adalah sekumpulan objek yang berbeda
- Setiap elemen atau objek didalam sebuah set / himpunan tidak boleh sama
- Bersifat Unordered
- Penyisipan, penghapusan, pencarian membutuhkan kompleksitas rata-rata O(1)

### 1.1 Cara membuat set

#### Manual

Kita dapat membuat set secara manual menggunakan list.

Namun dalam prakteknya, sebuah set memiliki beberapa aturan sehingga kita harus membuat fungsi fungsi agar set memenuhi segala aturannya

```python
def set_append(Set,Elemen) :
    if Elemen not in Set :
        Set.append(Elemen)
    return

def set_update(Set,Elemen) :
    for i in Elemen :
        set_append(Set,i)
    return

def set_pop(Set,Elemen) :
    for i in range(len(Set)) :
        if Set[i] == Elemen :
            Set.pop(i)
            return
    return

def set_cardinality(Set) :
    return len(Set)

def set_is_subset(Set1,Set2) :
    for i in Set1 :
        if i not in Set2 :
            return False
    return True

def set_is_superset(Set1,Set2) :
    for i in Set2 :
        if i not in Set1 :
            return False
    return True
    
```

```python
A = []

set_update(A,"Matematika Diskrit")

print(A)
```

#### Class Set

## 2. Operasi Himpunan

### 2.1 Union

Adalah operasi himpunan yang digunakan untuk menggabungkan seluruh elemen dari dua himpunan menjadi satu himpunan baru.

```python
X = {12,5,7,3}
Y = {2,3,5,2}
Z = {9,8,4,1}

print(X|Y)
print(X.union(Y))
```

### 2.2 Intersection

Adalah operasi yang menghasilkan himpunan baru yang berisi elemen-elemen yang sama (irisan) antara dua himpunan.

```python
print(X&Y)
X.intersection(Y)
```

### 2.3 Difference

adalah operasi yang menghasilkan himpunan berisi elemen yang ada di himpunan A tetapi tidak ada di himpunan B.

```python
print(Y - X)
```

### 2.4 Symmetric Difference

adalah operasi yang menghasilkan himpunan berisi elemen-elemen yang hanya muncul di salah satu himpunan, tetapi tidak di keduanya sekaligus.

```python
print(Y ^ Z)
```

### 2.5 Disjoint Set 

adalah kondisi di mana dua himpunan tidak memiliki elemen yang sama sama sekali. Disjoint set bukan operasi tapi relasi antar 2 himpunan.

```python
print(X.isdisjoint(Y))
print(Z.isdisjoint(Y))
```

## 5. Cardinality

Cardinality atau Kardinalitas menyatakan jumlah elemen dalam sebuah set.

Pada python, Cardinality dapat dicari menggunakan fungsi `len()`

```python
set_angka = {1,2,3,4,5,6,7,8,9}
```

```python
print(f"Jumlah Elemen Set Angka = {len(set_angka)}")
```

## 6. Komplemen

Komplemen merupakan seluruh anggota semesta yang tidak termasuk kedalam set yang dipilih.

Cara mencari komplemen dapat menggunakan Operasi `-`

```python
semesta = {1,2,3,4,5,6,7,8,9}
set_angka = {4,5,6}

set_angka_komplemen = semesta - set_angka

print(set_angka_komplemen)
```
