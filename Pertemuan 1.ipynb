## Set / Himpunan

- Himpunan (set) adalah sekumpulan objek yang berbeda

- Setiap elemen atau objek didalam sebuah set / himpunan tidak boleh sama

### Cara membuat set

#### Manual

Kita dapat membuat set secara manual menggunakan list.

Namun dalam prakteknya, sebuah set memiliki beberapa aturan sehingga kita harus membuat fungsi fungsi agar set memenuhi segala aturannya

```python
def set_append(Set,Elemen) :
    if Elemen not in Set :
        Set.append(Elemen)
    return

def set_create(Set,Elemen) :
    for i in Elemen :
        set_append(Set,i)
    return

def set_pop(Set,Elemen) :
    for i in range(len(Set)) :
        if Set[i] == Elemen :
            Set.pop(i)
            return
    return

def set_kardinalitas(Set) :
    return len(Set)

def set_is_subset(Set1,Set2) :
    for i in Set1 :
        if i not in Set2 :
            return False
    return True

def set_is_superset(Set1,Set2) :
    for i in Set2 :
        if i not in Set1 :
            return False
    return True
    
```

```python
A = []

set_create(A,"Matematika Diskrit")

print(A)
```

#### Class Set

Kita juga dapat membuat set dengan Class set karena didalam python sudah terdapat Class set yang siap diguunakan.

```python
set1 = set()        # Set Kosong
print(f"Set 1 = {set1}")

buah = ["Mangga", "Pisang", "Apel", "Pisang"]

set2 = set(buah)    # Set dengan list
print(f"Set 2 = {set2}")

set3 = {1,2,3}      # Set dengan bracket
print(f"Set 3 = {set3}")
```

### Cara menambah elemen set

#### add

Menambahkan 1 elemen

```python
set2.add("Anggur")
print(set2)
```

#### update

Menambahkan 1 atau banyak elemen menggunakan list

```python
set2.update(["Jambu","Pisang","Jeruk"])
print(set2)
```

### Cara menghapus elemen set

#### discard

Menghapus elemen tanpa pesan error (disarankan)

```python
set2.discard("Pepaya")
print(set2)

set2.discard("Pisang")
print(set2)
```
