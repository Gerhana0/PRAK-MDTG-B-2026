## Set / Himpunan

- Himpunan (set) adalah sekumpulan objek yang berbeda

- Setiap elemen atau objek didalam sebuah set / himpunan tidak boleh sama

### Cara membuat set

#### Manual

Kita dapat membuat set secara manual menggunakan list.

Namun dalam prakteknya, sebuah set memiliki beberapa aturan sehingga kita harus membuat fungsi fungsi agar set memenuhi segala aturannya

```python
def set_append(Set,Elemen) :
    if Elemen not in Set :
        Set.append(Elemen)
    return

def set_update(Set,Elemen) :
    for i in Elemen :
        set_append(Set,i)
    return

def set_pop(Set,Elemen) :
    for i in range(len(Set)) :
        if Set[i] == Elemen :
            Set.pop(i)
            return
    return

def set_cardinality(Set) :
    return len(Set)

def set_is_subset(Set1,Set2) :
    for i in Set1 :
        if i not in Set2 :
            return False
    return True

def set_is_superset(Set1,Set2) :
    for i in Set2 :
        if i not in Set1 :
            return False
    return True
    
```

```python
A = []

set_update(A,"Matematika Diskrit")

print(A)
```

#### Class Set

Kita juga dapat membuat set dengan Class set karena didalam python sudah terdapat Class set yang siap diguunakan.

```python
set1 = set()        # Set Kosong
print(f"Set 1 = {set1}")

buah = ["Mangga", "Pisang", "Apel", "Pisang"]

set2 = set(buah)    # Set dengan list
print(f"Set 2 = {set2}")

set3 = {1,2,3}      # Set dengan bracket
print(f"Set 3 = {set3}")
```

## Superset dan Subset

```python
set_a = {1,2,3,4,5,6,7}
set_b = {3,4,5}
set_c = {5,6}
```

### Superset

#### Metode issuperset

```python
set_a.issuperset(set_b)
```

#### Operasi `>=`

```python
set_a >= set_b
```

```python
set_b >= set_c
```

### Subset

#### Metode issubset

```python
set_b.issubset(set_a)
```

#### Operasi `<=`

```python
set_b <= set_a
```

```python
set_a <= set_b
```

## Membership

Membership menyatakan ada atau tidaknya sebuah elemen didalam suatu set

Dalam python, Membership dapat di cek menggunakan `in` atau `not in`

```python
sayuran = {"Kubis","Selada","Bayam","Sawi"}
```

```python
"Tomat" in sayuran
```

```python
"Tomat" not in sayuran
```

## Cardinality

Cardinality atau Kardinalitas menyatakan jumlah elemen dalam sebuah set

Pada python, Cardinality dapat dicari menggunakan fungsi `len()`

```python
set_angka = {1,2,3,4,5,6,7,8,9}
```

```python
print(f"Jumlah Elemen Set Angka = {len(set_angka)}")
```

## Komplemen

Komplemen merupakan seluruh anggota semesta yang tidak termasuk kedalam set yang dipilih.

Cara mencari komplemen dapat menggunakan Operasi `-`

```python
semesta = {1,2,3,4,5,6,7,8,9}
set_angka = {4,5,6}

set_angka_komplemen = semesta - set_angka

print(set_angka_komplemen)
```
