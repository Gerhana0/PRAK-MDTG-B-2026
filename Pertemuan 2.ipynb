## 1. Visualisasi Relasi

```python
import matplotlib.pyplot as plt
import numpy as np

def visualisasi_relasi (Set1,Set2,Relasi={},Judul="",NamaSet1="",NamaSet2="",WarnaSet1="skyblue",WarnaSet2="lightgreen") :
    fig, ax = plt.subplots(figsize=(8, 6))
    pos_x_Set1 = np.zeros(len(Set1))
    pos_x_Set2 = np.ones(len(Set2))
    pos_y_Set1 = np.linspace(1,5,len(Set1))
    pos_y_Set2 = np.linspace(1,5,len(Set2))
    ax.scatter(pos_x_Set1, pos_y_Set1, s=500, c=WarnaSet1, edgecolors='black', zorder=2)
    ax.scatter(pos_x_Set2, pos_y_Set2, s=500, c=WarnaSet2, edgecolors='black', zorder=2)
    koor_Set1 = {}
    koor_Set2 = {}
    
    indeks = 0
    for i in Set1:
        ax.text(-0.1, pos_y_Set1[indeks], str(i), fontsize=12, ha='right', va='center', fontweight='bold')
        koor_Set1[str(i)] = (0,pos_y_Set1[indeks])
        indeks += 1
        
    indeks = 0
    for i in Set2:
        ax.text(1.1, pos_y_Set2[indeks], str(i), fontsize=12, ha='left', va='center', fontweight='bold')
        koor_Set2[str(i)] = (1,pos_y_Set2[indeks])
        indeks += 1
        
    for i in Relasi:
        for j in Relasi[i]:
            if i in Set1 and j in Set2 :
                ax.annotate("", 
                            xy=koor_Set2[str(j)],    
                            xytext=koor_Set1[str(i)], 
                            arrowprops=dict(arrowstyle="-", connectionstyle="arc3", color='black', lw=1.5),
                            zorder=1)
                
    ax.set_title(Judul, fontsize=16, pad=20)
    ax.set_xticks([0, 1])
    ax.set_xticklabels([NamaSet1, NamaSet2], fontsize=14, fontweight='bold')
    ax.set_yticks([])
    ax.set_xlim(-0.5, 1.5)
    ax.set_ylim(0, 6)
    
    plt.tight_layout()
    plt.show()
    
```

```python
Asprak = {"Elias","Rexy","Alvin","Hana"}
Prodi = {"Sains Data", "Informatika"}
Relasi = {
    "Elias" : ["Sains Data"],
    "Rexy" : ["Sains Data"],
    "Alvin" : ["Sains Data"],
    "Hana" : ["Informatika"]
}

visualisasi_relasi(Set1=Asprak , Set2=Prodi , Relasi=Relasi , NamaSet1="Asprak" , NamaSet2="Prodi", Judul="Prodinya Para Asprak Matdis")
```
